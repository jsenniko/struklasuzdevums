## Uzdevums

Ūdens strūkla iztek no šļūtenes ar konstantu nezināmu ātrumu *v*. Šļūtene tiek patvaļīgi rotēta fiksētā *x-y* plaknē, tās gals tiek fiksēti noturēt pozīcijā $x=y=0 m$. Šļūtenes izplūdes leņķis pret horizontu ir lielāks par $45^\circ$. Katrā laika momentā ūdens strūkla gaisā ieņem noteiktu formu. Dotajā attēlā ir parādīta strūklas formā kāda laika momentā. Lietojot strūklas formu no šī attēla, noteikt ūdens ātrumu, ar kuru ūdens izplūst no šļūtenes. Pieņemt, ka brīvās krišanas paātrinājums ir $9.8 m/s^2$

### Risinājums

Šajā izklājloksnē papildus uzdevuma pamatjautājumam - noteikt izplūdes ūdens ātrumu - tiks aprēķināta arī ūdens strūklas forma laikā. Tāpat arī šļūtenes leņķa atkarība no laika. Tiks izveidotas strūklas pārvietošanās animācijas

#### Pieņēmumi
Uzskatīsim, ka ūdensstrūklas punkti ir neatkarīgi, katrs no tiem pēc izplūšanas no šļūtenes gala pārvietojas Zemes gravitācijas laukā. Gaisa pretestību neņemsim vērā.

#### Punkta pārvietošanās trajektorija gravitācijas laukā
To var izteikt ar sekojošu divu, no laika $t$ atkarīgu, vienādojumu palīdzību. Tie apraksta pārvietošanos $x$ un $y$ virzienā, brīvās krišanas paātrinājums vērsts negatīvā *y* virzienā. $\alpha$ ir leņķis pret horizontu, ar kuru dotais strūklas punkts izplūdis no caurules laika momentā $t=0$

$$\begin{cases} x=v \cos\alpha \cdot t \\ y=v \sin\alpha\cdot t -\frac{g t^2}{2}  \end{cases}$$

Izslēdzot t no pārvietošanās vienādojumiem, iegūst sakarību starp koordinātēm $x$ un $y$:
$$
 t=\frac{x}{v cos(\alpha)} \\
 y = x \tan\alpha - \frac{g x^2}{2 v^2 \cos^2(\alpha)} \\
 \frac{1}{\cos^2\alpha}=1+\tan^2\alpha \\
 y=x \tan\alpha  - \frac{g x^2}{2 v^2}(1+\tan^2\alpha)
$$

Katram x pie dažādiem izviešanas leņķiem ir dažādi y. Pie tam $y$ ir atkarīgs no $x$, izviešanas ātruma $v$ un leņķa $\alpha$  
$y=y(x,v^2, \tan\alpha)$ 

#### Apliecējas līknes noteikšana


Atradīsim $y$ maksimālo vērtību atkarībā no $x$: $y_{max} = y_{max}(x, v^2)$, šo līkni sauksim par *apliecēju*. Lai to iegūtu, 
atvasināsim y izteiksmi pēc $\tan\alpha$ un pielīdzinot atvasinājumu 0 noteiksim $\tan\alpha$ izteiksmi maksimālajam y. Ievietojot atpakaļ $y$ izteiksmē iegūsim sakarību $y_{max} = y_{max}(x, v^2)$:
$$
 \frac{dy}{d(\tan\alpha))}=x-\frac{g x^2}{v^2}\tan\alpha = 0 \\
 \tan\alpha =\frac{v^2}{g x} \\
 y_{max} = \frac{v^2}{2g} - \frac{gx^2}{2v^2}
$$

#### Izsviešanas ātruma noteikšana  
$y_{max}$ sakarību var izmantot ātruma $v$ noteikšanai. Visiem uzdotās strūklas momentuzņēmuma punktiem no šis sakarības atrodam $v=v(x_{i}, y_{i})$. Meklētais ātrums ir maksimālais no visām vērtībām, t.i. uzdotā strūklas līnija nešķērso nevienu no potenciālajām apliecējām, bet tikai tai pieskaras no lejpuses.  
Sastādām kvadrātvienādojumu pret $\frac{v^2}{g}$:
$$
 \left( \frac{v^2}{g} \right) ^2 - 2y\frac{v^2}{g} - x^2 = 0 \\ 
$$

Tā reālais atrisinājums ir:
$$ v=\sqrt{ g \left( y+\sqrt{(x^2+y^2)} \right)}$$


#### Katra punkta izplūšanas laika momenta noteikšana

Pēc tam, kad ir noteikts $v$, varam noteikt katram no uzdotās strūklas punktiem tam atbilstošo sākummonentu un izplūšanas leņķi. Tādam nolūkam no kustības vienādojumiem izslēdzam leņķi:
$$
 (vt)^2 = \left( v \cos\alpha \cdot t \right)^2+\left( v \sin\alpha \cdot t \right)^2 = x^2+\left(y+\frac{g t^2}{2}\right)^2
$$
No šī vienādojuma sastādam kvadrātvienādojumu priekš $t^2$:
$$
 \frac{g^2}{4}\left(t^2\right)^2 +(gy-v^2)t^2+x^2+y^2=0
$$
Šim kvadrātvienādojumam ir divi atrisinājumi priekš $t^2$:
$$
 t^2_{1,2} = \frac{v^2-g y \pm \sqrt{(gy-v^2)^2-g^2(x^2+y^2))}}{\frac{g^2}{2}}
$$
Izplūšanas leņki nosakām no sakarības:
$$
 \alpha=\arccos\frac{x}{vt}
$$

#### Strūklas punktu laika momentu sakārtošana
Katram no strūklas momentuzņēmuma punktiem jāatrod abas iespējamā laika vērtības (šis laiks katram strūklas punktam ir laiks, ko punkts atradies gaisā, sākot ar tā izplūšanas laiku, t.i., lielākas t vērtības ir punktiem, kas izplūduši agrāk). Tālāk jāizvēlas tāda laika vērtību secība (ņemot pirmo vai otro no t risinājumiem), kurā laiks gar strūklas punktiem monotoni pieaugs. Tādējādi būsim arī ieguvuši strūklas izplūdes leņķa maiņas laikā grafiku.

### Trajektorijas aprēķina funkcijas

```python
# trajektorijas aprēķina funkcijas
g=9.8
def pozicija_x(alpha, deltaT, v):
    return np.cos(np.radians(alpha))*v*deltaT
def pozicija_y(alpha, deltaT, v):
    return np.sin(np.radians(alpha))*v*deltaT-g/2*deltaT*deltaT
v_pozicija_x=pozicija_x#np.vectorize(pozicija_x)
v_pozicija_y=pozicija_y#np.vectorize(pozicija_y)
```

### Izsviešanas ātruma aprēķins

Aprēķinam $v_{max}$ vērtības katram no uzdotajiem punktiem. Mainīgais *vmax* ir masīvs, kas satur tik pat elementus, cik uzdotiem masīvi *x0* un *y0*

*vmod* ir izviešanas ātrums - oriģinālā uzdevuma atrisinājums. Veicam arī pārbaudi, vai laika aprēķina vienādojuma diskriminanta nav mazāks par 0. Ja tas ir tā, tad uzdotajam strūklas momentuzņēmums nevar būt korekts atbilstoši uzdevuma nosacījumiem

```python
# ātrumu atrodam no punkta uz tādas apliecējas, kurā ir max ātrums
vmax=np.sqrt(g*(y0+np.sqrt(x0*x0+y0*y0)))
# noapaļošanas problēma vēlāk, tāpec pievienoju  mazo skaitli
vmod=np.max(vmax)+0.000000001
diskrim=(g*y0-vmod*vmod)**2-g*g*(x0*x0+y0*y0)
# kontrolpārbaude, vai diskrimintants nav <0
#diskrim[diskrim<0.0], vmod
if any(diskrim<0.0):
    print('Nekorekts atrisinājums!!!!')
```

### Trajektorijas aprēķins

Aprēķinam laiku un leņķi katram nociparotajam strūklas punktam

Eksistē **divi** atrisinājumi katram no punktiem ar dažādiem izsviešanas laikiem un leņķiem. Ievietosim abus atrisinājumus *pandas* tabulā

```python
# laika un leņķa aprēķins katram nociparotajam strūklas punktam
# eksistē divi atrisinājumi katrā punktā ar dažādiem laikiem un leņķiem
dt1=np.sqrt((vmod*vmod-g*y0 - np.sqrt((g*y0-vmod*vmod)**2-g*g*(x0*x0+y0*y0)))/(g*g/2))
dt2=np.sqrt((vmod*vmod-g*y0 + np.sqrt((g*y0-vmod*vmod)**2-g*g*(x0*x0+y0*y0)))/(g*g/2))
alpha0_1=np.degrees(np.arccos(x0/vmod/dt1))
alpha0_2=np.degrees(np.arccos(x0/vmod/dt2))
df=pd.DataFrame({"t1":dt1,"alpha1":alpha0_1, "t2":dt2,"alpha2":alpha0_2, "x0":x0,"y0":y0})
#HTML(df.to_html())
```

#### Trajektorijas punktu sakārtošana laika pieaugšanas secībā
Veiksim to manuāli. Tā kā momentuzņēmuma ciparošana ir aptuvena, var gadīties, ka nevar atrast secīgus punktus, kuros pieaug laiks. Šādos gadījumos jāpārbauda ciparošana, vai jāveic laika korekcija.

Mainīgajā *traj_secība* katram no uzdotajiem punktiem uzdosim, vai to ņem no atrisinājuma 1 vai 2.

Tabulā *df_traj* uzkrāsim

```python
# pamata uzdošana - punkti 0-30 ->1, punkti 40,68 ->2
traj_seciba=[(i,1) for i in range(40)] +[(i,2) for i in range(40,69)]
```

```python
def convertTraj(traj):
    df_traj={}
    traj_tim=[]
    traj_alpha=[]
    traj_x=[]
    traj_y=[]
    for i in range(len(traj)):
        d=df.loc[traj[i][0]]
        traj_tim.append(d['t1'] if traj[i][1]==1 else d['t2'])
        traj_alpha.append(d['alpha1'] if traj[i][1]==1 else d['alpha2'])
        traj_x.append(d['x0'])
        traj_y.append(d['x0'])
    df_traj['x'],df_traj['y'],df_traj['t'],df_traj['a'] = traj_x, traj_y, traj_tim, traj_alpha
    df_traj = pd.DataFrame(df_traj)
    return df_traj
df_traj = convertTraj(traj_seciba)
# nulltais leņķis - sākumpunkts
df_traj['a'].iloc[0]=88.0
```

#### Funkcija, kas var pārinterpolēt leņķus no uzdotajiem punktiem uz jebkuru laika momentu
Izmantojam 3-kārtas splainu interpolāciju

```python
def interpolet_lenkus(times, alphas, timescalc):
    finterp=scipy.interpolate.interp1d(times, alphas, kind='cubic')
    alphascalc=finterp(timescalc)    
    return alphascalc
```

#### Trajektorijas saskaņošana   
Šajā attēlā uzzīmejam leņķus atkarībā no laika (no mainīgā *df_traj*). Ja interpolācijas līkne nav gluda, vai dažos punktos secīgi laiks samazinās, jāveic manuālas korekcijas. Šeit tas darīts punktam ar numuru 39.

```python
df_traj_korig=df_traj.copy()
# jākoriģē 39 punkts, lai animācijā nebūtu lielu lauzumu
uuu=(df_traj_korig.loc[40]-df_traj_korig.loc[38])
df_traj_korig.iloc[39]['t']= (df_traj_korig.loc[39]['a']-df_traj_korig.loc[38]['a'])/uuu['a']*uuu['t']+df_traj_korig.loc[38]['t']
```

```python
ddf = df_traj_korig
# trajektoriju leņķis no laika
# ievērojam, ka ap 1.5 sekundi vienam no punktiem laiks iet atpakaļ - to pēc tam jākoriģē
ncalcpoints=15000
tstepcalc=df_traj_korig['t'].max()/ncalcpoints
timescalc=np.arange(ncalcpoints)*tstepcalc
alphascalc=interpolet_lenkus(ddf['t'], ddf['a'], timescalc)

fig, ax=plt.subplots(figsize=(20,10))
ax.plot(df_traj['t'],df_traj['a'], marker='o', label='nekoriģēts')
ax.plot(df_traj_korig['t'],df_traj_korig['a'], marker='o', label='koriģēts')
ax.plot(timescalc,alphascalc, label='interpolēts')
ax.grid()
ax.plot([0,2.5],[45,45],color='black',ls=':',lw=2)
ax.legend(loc='best')
ax.set_xlabel('Laiks, s')
ax.set_ylabel('Leņķis, $^\circ$')
plt.show()
```

### Animācijas kadri

```python
# trajektoriju fragmenti
show_tails=False
# punkti
show_points=True
# rezultātu kadru katalogs
animation_directory='animation'
if not os.path.exists(animation_directory):
    os.makedirs(animation_directory)
```

```python
def calc_position(frametime):    
    #frametime=frametimes[iframe]
    alpha0_t=np.interp(frametime,launctimescalc,alphascalc)
    # laiks, cik ilgi lidojis katrs no izvēlētajiem punktiem
    flighttimes=np.where(frametime-launctimescalc>0,frametime-launctimescalc,0)
    # gala  pozīcijas aprēķins
    pxcalc=v_pozicija_x(alphascalc, flighttimes, vmod)
    pycalc=v_pozicija_y(alphascalc, flighttimes, vmod)
    
    return alpha0_t, pxcalc, pycalc
```

### Animācija ar matplotlib līdzekļiem

```python
import matplotlib.animation
```

```python
def draw_position(frametime, alpha0_t, pxcalc, pycalc):
    drawline.set_data(pxcalc,pycalc)
    if show_points:
        drawpnts.set_data(pxcalc[::300],pycalc[::300])
    
    anot1.set_text(f'$t = {frametime:.2f} s$')
    if frametime<maxt:
        anot2.set_text(fr'$\alpha = {alpha0_t:.2f}^\circ$')    
```

```python
# ja grib noskatīties, tad %matplotlib notebook, ja tikai noglabāt tad labāk %matplotlib inline
%matplotlib notebook
matplotlib.rcParams.update({'font.size': 20})
def update(frametime):
    alpha0_t, pxcalc, pycalc = calc_position(frametime)
    draw_position(frametime, alpha0_t, pxcalc, pycalc)   
    
fig, ax=plt.subplots(figsize=(12,7))
ax.set_aspect(1)
drawline, = ax.plot([],[], color='blue',lw=4)

# punkti
if show_points:
    drawpnts, =ax.plot([],[], color='blue',lw=0, marker='o', ms=10)


# apliecēja
ax.plot(xmax,ymax, color='black',ls='--',lw=2)
ax.set_ylim(0,8)
ax.set_xlim(0,15)
ax.set_xlabel('X, m')
ax.set_ylabel('Y, m')
ax.grid()
anot1 = ax.annotate('',(12.5,7.5))
anot2 = ax.annotate(r'',(12.5,7.1))

anim = matplotlib.animation.FuncAnimation(fig, update, frames=frametimes, repeat=False)

plt.show()    
```

### Animācijas kadru izgatavošana failos

```python
%matplotlib inline
matplotlib.rcParams.update({'font.size': 20})
for iframe, frametime in enumerate(frametimes):
    
    alpha0_t, pxcalc, pycalc = calc_position(frametime)
    
    fig, ax=plt.subplots(figsize=(19,9.99))
    
    if show_tails:
        # trajektoriju fragmenti    
        dttraj=0.01
        dttrajback=0.3
        for i in range(0,len(alphascalc),300):
            times1=np.arange(flighttimes[i],max(0,flighttimes[i]-dttrajback),-dttraj)
            px1=v_pozicija_x(alphascalc[i], times1, vmod)
            py1=v_pozicija_y(alphascalc[i], times1, vmod)
            ax.plot(px1,py1, ls='--',lw=1, color='gray')
    
    
    # līnija
    ax.plot(pxcalc,pycalc, color='blue',lw=4)
    # punkti
    if show_points:
        ax.plot(pxcalc[::300],pycalc[::300], color='blue',lw=0, marker='o', ms=10)
    
    
    # apliecēja
    ax.plot(xmax,ymax, color='black',ls='--',lw=2)
    ax.set_ylim(0,8)
    ax.set_xlim(0,15)
    ax.set_xlabel('X, m')
    ax.set_ylabel('Y, m')
    ax.grid()
    
    
    
    ax.annotate('$t = {:.2f} s$'.format(frametime),(12.5,7.5))
    if frametime<maxt:
        ax.annotate(r'$\alpha = {:.2f}^\circ$'.format(alpha0_t),(12.5,7.1))
    fig.savefig(f'{animation_directory}\\{iframe:04d}.jpg',bbox_inches='tight', dpi=124.3)
    plt.close()
    #plt.show()
    
```
